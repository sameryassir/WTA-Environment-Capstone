```python
import pandas as pd

# Load the datasets
catapult = pd.read_csv("catapult environmental conditions data export.csv")
smartabase = pd.read_csv("sb environmental conditions data export.csv")
matching_table = pd.read_csv("wta_tournament_matchingtable_bt.csv")

# Display the first few rows of each to confirm structure
print("Catapult:")
display(catapult.head())

print("Smartabase:")
display(smartabase.head())

print("Tournament Matching Table:")
display(matching_table.head())
```

```python
# Standardize environmental metric column names
catapult_renamed = catapult.rename(columns={
    "Heat Stress Index": "WBGT",
    "Heat Rule In/Out": "EWCR"
})

smartabase_renamed = smartabase.copy()  # Already uses WBGT, AQI Reading, and EWCR

# Ensure both datasets have matching variable names
common_columns = ["Tournament Name", "Year", "Date", "Time", "WBGT", "AQI Reading", "EWCR"]

# If there are formatting differences in column names (e.g., spaces), we will clean them later
```

```python
# Select and rename relevant columns from Catapult
catapult_cleaned = catapult[[
    "Heat Stress Tournament", "Years", "Heat Stress Date",
    "Time Of Heat Reading", "Heat Stress Index", "AQI Reading",
    "Heat Rule In/out", "Air Temperature", "Relative Humidity", "Type Of Day"
]].rename(columns={
    "Heat Stress Tournament": "Tournament Name",
    "Years": "Year",
    "Heat Stress Date": "Date",
    "Time Of Heat Reading": "Time",
    "Heat Stress Index": "WBGT",
    "Heat Rule In/out": "EWCR",
    "Air Temperature": "Air Temperature",
    "Relative Humidity": "Relative Humidity",
    "Type Of Day": "Type of Day"
})
```

```python
# Select and rename relevant columns from Smartabase
smartabase_cleaned = smartabase[[
    "Tournament Name", "Tournament Year", "Log Date", "Time of Reading",
    "WBGT", "AQI Reading", "EWCR", "Air Temperature (TA)",
    "Relative Humidity", "Type of Day"
]].rename(columns={
    "Tournament Year": "Year",
    "Log Date": "Date",
    "Time of Reading": "Time",
    "Air Temperature (TA)": "Air Temperature"
})
```

```python
# Strip and lower-case for easy matching
catapult_cleaned["EWCR"] = catapult_cleaned["EWCR"].str.strip().str.upper()
smartabase_cleaned["EWCR"] = smartabase_cleaned["EWCR"].str.strip().str.upper()

# Normalize catapult EWCR values to match Smartabase format
catapult_cleaned["EWCR"] = catapult_cleaned["EWCR"].replace({
    "IN": "IN – MODIFICATION OF PLAY",
    "OUT": "OUT"
})
```

```python
# Merge both datasets (row-wise concatenation)
merged_env_data = pd.concat([catapult_cleaned, smartabase_cleaned], ignore_index=True)

# Confirm merge success
print("Merged environmental dataset shape:", merged_env_data.shape)
display(merged_env_data[:20])
```

```python
import pandas as pd
import re

def extract_tournament_info(tournament_name):
    if isinstance(tournament_name, str):
        match = re.match(r"^(\d{4})\s*-\s*(.+?)(?:\s+(\d{3,4}))?$", tournament_name.strip())
        if match:
            year = int(match.group(1))
            name = match.group(2).strip()
            level = match.group(3).strip() if match.group(3) else None
            return pd.Series([year, name, level])
    return pd.Series([None, tournament_name, None])  # fallback for NaN or non-string

# Apply safely to extract year, city, and level
merged_env_data[["Extracted Year", "Tournament City", "Tournament Level"]] = merged_env_data["Tournament Name"].apply(extract_tournament_info)

# Fill missing years where applicable
merged_env_data["Year"] = merged_env_data["Year"].fillna(merged_env_data["Extracted Year"])

# Drop temporary helper column
merged_env_data.drop(columns=["Extracted Year"], inplace=True)

# Preview the cleaned data
display(merged_env_data[["Tournament Name", "Year", "Tournament City", "Tournament Level"]][:10000])
```

```python
# Make a clean version of the tournament city column by removing levels like 125, 250, etc.
merged_env_data["Cleaned City"] = (
    merged_env_data["Tournament City"]
    .astype(str)
    .str.replace(r"\s*\d{3,4}(?:\s*#\d+)?", "", regex=True)
    .str.replace(r"#\d+", "", regex=True)
    .str.strip()
)
```

```python
# Strip off " GS" if it appears at the end
merged_env_data["Cleaned City"] = merged_env_data["Cleaned City"].str.replace(r"\s+GS$", "", regex=True)
```

```python
# Use a function to map non-city tournaments based on year
def map_special_city(row):
    name = row["Cleaned City"]
    year = row["Year"]

    if name == "WTA FINALS":
        return {
            2021: "GUADALAJARA",
            2022: "FORT WORTH",
            2023: "CANCUN"
        }.get(year, "UNKNOWN")  # fallback to UNKNOWN

    if name == "OLYMPICS":
        return {
            2021: "TOKYO",
            2024: "Paris"
        }.get(year, "UNKNOWN")  # update for future Olympics as needed

    if name == "UNITED CUP":
        return "SYDNEY"

    if name in ["US OPEN", "US OPEN GS"]:
        return "NEW YORK"
    
    if name in ["WIMBLEDON", "WIMBLEDON GS"]:
        return "LONDON"
    
    if name in ["ROLAND GARROS", "ROLAND GARROS GS"]:
        return "PARIS"
    
    if name == "AUSTRALIAN OPEN":
        return "MELBOURNE"

    return name  # default fallback is the cleaned city itself

# Apply the function to create the mapped city column
merged_env_data["Mapped City"] = merged_env_data.apply(map_special_city, axis=1)
```

```python
# See a few unique results to confirm mapping worked
display(
    merged_env_data[["Tournament City", "Cleaned City", "Mapped City", "Year"]]
    .drop_duplicates()
    .sort_values("Mapped City")
    .head(20)
)
```

```python
import pandas as pd

# List of your CSV files
file_paths = [
    "data/Merged_WTA_Env_with_region.csv",
    "data/Merged_WTA_Env.csv"
]

for file in file_paths:
    print(f"Processing: {file}")
    
    # Load the dataset
    df = pd.read_csv(file)

    # Replace 'Tournament City' with values from 'Mapped City'
    df['Tournament City'] = df['Mapped City']

    # Drop the extra columns
    df.drop(columns=['Cleaned City', 'Mapped City'], inplace=True, errors='ignore')

    # Save the updated version
    df.to_csv(file, index=False)
    print(f"Updated and saved: {file}\n")
```

```python
# Reiterate through the same files to drop any 'Unnamed' columns
for file in file_paths:
    print(f"🧽 Cleaning unnamed columns in: {file}")
    
    df = pd.read_csv(file)

    # Drop columns starting with 'Unnamed'
    unnamed_cols = [col for col in df.columns if col.startswith("Unnamed")]
    df.drop(columns=unnamed_cols, inplace=True)

    # Save the cleaned data
    df.to_csv(file, index=False)
    print(f"✅ Dropped {len(unnamed_cols)} unnamed columns from: {file}\n")
```
