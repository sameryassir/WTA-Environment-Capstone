```python
# Import necessary libraries
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Check what files exist in the current directory
print("Files in the current directory:", os.listdir())

# Check if the files exist in the expected path
sb_path = "sb environmental conditions data export.csv"
catapult_path = "catapult environmental conditions data export.csv"
matching_table_path = "wta_tournament_matchingtable_bt.csv"

print(f"Smartabase Exists: {os.path.exists(sb_path)}")
print(f"Catapult Exists: {os.path.exists(catapult_path)}")
print(f"Matching Table Exists: {os.path.exists(matching_table_path)}")
```

```python
# Load datasets
sb_data = pd.read_csv(sb_path)
catapult_data = pd.read_csv(catapult_path)
matching_table = pd.read_csv(matching_table_path)

# Display first few rows
print("âœ… Smartabase Data Sample:")
display(sb_data.head())

print("\nâœ… Catapult Data Sample:")
display(catapult_data.head())

print("\nâœ… Tournament Matching Table Sample:")
display(matching_table.head())

# Show dataset shapes
print(f"\nSmartabase Shape: {sb_data.shape}")
print(f"Catapult Shape: {catapult_data.shape}")
print(f"Matching Table Shape: {matching_table.shape}")
```

```python
# Standardize date format
sb_data["Date"] = pd.to_datetime(sb_data["Date"], errors='coerce')
catapult_data["Date"] = pd.to_datetime(catapult_data["Date"], errors='coerce')

# Check date format conversion
print("\nâœ… Smartabase Date Format:")
print(sb_data["Date"].head())

print("\nâœ… Catapult Date Format:")
print(catapult_data["Date"].head())
```

```python
# Rename columns for consistency in merging
sb_data.rename(columns={"Tournament Name": "Tournament", "Tournament Year": "Year"}, inplace=True)
catapult_data.rename(columns={"Heat Stress Tournament": "Tournament", "Years": "Year"}, inplace=True)

# Verify updated column names
print("\nâœ… Updated Smartabase Columns:", sb_data.columns.tolist())
print("\nâœ… Updated Catapult Columns:", catapult_data.columns.tolist())
```

```python
# Rename the duplicate "Year" column in Smartabase
sb_data.rename(columns={"Year": "Year_sb"}, inplace=True)

# Rename the duplicate "Year" column in Catapult
catapult_data.rename(columns={"Year": "Year_catapult"}, inplace=True)

# Verify the updated column names
print("\nâœ… Updated Smartabase Columns:", sb_data.columns.tolist())
print("\nâœ… Updated Catapult Columns:", catapult_data.columns.tolist())
```

```python
merged_data = sb_data.merge(
    catapult_data, 
    on="Tournament", 
    how="outer", 
    suffixes=("_sb", "_catapult")
)

print("\nâœ… Merged Data Sample:")
display(merged_data.head())

# Check if merge was successful
print(f"\nTotal Rows: {merged_data.shape[0]}, Unique Tournaments: {merged_data['Tournament'].nunique()}")
```
